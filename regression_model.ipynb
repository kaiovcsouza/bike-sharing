## Avaliação dos modelos

```python
x = bikes.drop('cnt', axis=1).values
y = bikes['cnt'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

RANDOM_STATE=42

def build_regressors():
    regressors = []

    regressors.append(
        ('lr', 
        LinearRegression(), 
        {'fit_intercept':[True, False]})
    )

    regressors.append(
        ('svr', 
        SVR(), 
        {'C':[0.1,1,10],
        'gamma':[0.00001, 0.0001, 0.001, 0.01, 1],
        'epsilon':[0.0001 , 0.005, 0.001, 1, 5, 10]
        })
    )

    regressors.append(
        ('dt', 
        DecisionTreeRegressor(), 
        {'random_state':[RANDOM_STATE]})
    )

    regressors.append(
        ('rf', 
        RandomForestRegressor(), 
        {'n_estimators':[10],
        'random_state':[RANDOM_STATE]
        })
    )
    return regressors
```

```python
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error

train_results = []
test_results = []
names = []
scoring = 'neg_root_mean_squared_error'

regressors = build_regressors()
kfold = KFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
for name, regressor, params in regressors:
    print(f'>> regressor : {name}')
    # treinamento
    gs = GridSearchCV(regressor, param_grid=params, scoring=scoring, verbose=0)
    gs.fit(x_train, y_train)
    cv_results = cross_val_score(gs, x_train, y_train, cv=kfold, scoring=scoring)
    names.append(name)
    train_results.append(abs(cv_results))

    # teste
    y_pred = gs.predict(x_test)

    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    test_results.append(rmse)
```

```python
fig, ax = plt.subplots()
fig.suptitle('Desempenho em treinamento')
_=plt.boxplot(train_results)
ax.set_xticklabels(names)
ax.set_ylabel('RMSE')
ax.set_xlabel('Regressores')
```

```python
fig, ax = plt.subplots()
fig.suptitle('Desempenho em teste')
_=plt.bar(names, test_results)
ax.set_ylabel('RMSE')
ax.set_xlabel('Regressores')
```

### O modelo escolhido é o SVM, ele apresentou um score maior que os outros modelos
